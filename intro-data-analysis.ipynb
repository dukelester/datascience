```python
def leisure_cost(return_flight, hotel_per_day, weekly_car_rental,number_of_days,no_weeks):
    hotel_cost = hotel_per_day * number_of_days
    car_cost = weekly_car_rental * no_weeks
    
    total_trip_cost = return_flight + hotel_cost + car_cost
    
    return total_trip_cost

    
```

```python
def loan_emi(amount, duration,rate,down_payments):
    loan_amount = amount -down_payments
    emi = loan_amount / duration
    
    return emi

loans= [{ 'amount':828400,'duration': 120,'rate':0.11,'down_payment':100000},
        { 'amount':4633400,'duration': 240,'rate':0.06,'down_payment':0},
        { 'amount':42900,'duration': 90,'rate':0.08,'down_payment':8900},
        { 'amount':983000,'duration': 16,'rate':0.14,'down_payment':0},
        { 'amount':15230,'duration': 48,'rate':0.07,'down_payment':4300}]

values = []
for loan in loans:
    print(loan['amount'], loan['duration'], loan['rate'], loan['down_payment'])

    
    emi = loan_emi(loan['amount'],loan['duration'],loan['rate'], loan['down_payment'])  
    amount = round(loan['amount'])
    duration = round(loan['duration'])
    rate = round(loan['rate'])
    down_payment = round(loan['down_payment'])
    values.append(emi)
# print(values)

    file = open('loans.txt', mode='w')
    file.write('{},{}, {},{} \n'.format(amount,duration,rate,down_payment,emi))
file.close()

```
