```python
import matplotlib.pyplot as plt

plt.close('all')

ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/4/2021', periods=1000))
ts = ts.cumsum()
ts
ts.plot()
#the plot() is convinient in plotting all the columns and rows with labels
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=['A', 'B', 'C', 'D'])

df = df.cumsum()
plt.figure()
df.plot()
plt.legend(loc='best')
```
