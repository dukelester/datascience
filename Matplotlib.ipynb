MATPLOTLIB TUTORIAL

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
'''
Use numpy to generate random data

'''
import numpy as np
x = np.linspace(0,15,25)
print('the x values ---==',x)
y = x * x + 2
print()
print('The y values ==--',y)
#plotting the graph

pylab.plot(x,y, 'r') #r is the color = red

"""
Drawing sub graphs
"""
pylab.subplot(1,2,1) #the brackets we put rows, columns and indeces
pylab.plot(x,y, 'r--') #the line style is --
pylab.subplot(1,2,2)
pylab.plot(x,y,'g*-')



```

```python
from matplotlib import pyplot as plt

fig = plt.figure()
axis = fig.add_axes([0.5,0.1,0.8,0.8]) #control the figure
axis.plot(x,y,'lime')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)#rows and columns
for ax in axes:
    ax.plot(x,y,'red')
```

```python
fig = plt.figure()
#control the left, right ,width and height of the canvas.
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #big axis
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #the smaller axis

axes1.plot(x,y, 'r')
axes2.plot(y,x,'g')
plt.legend(['label1','label2'])
plt.xlabel('The X axis')
plt.ylabel('The Y axis')
```

```python
fig = plt.figure(figsize=(10,9),dpi=300)
fig.add_subplot()
plt.plot(x,y, 'r')
plt.legend(['graph'])
plt.xlabel('The x-axis')
plt.ylabel('The y-axis')
plt.title('My graph')
```

```python
fig, axes = plt.subplots(dpi=200)
axes.plot(x, x+1, color='red')
axes.plot(x, x+2, color='orangered')
axes.plot(x, x+3, color='blue')
```

```python
fig, axes = plt.subplots(dpi=200)
axes.plot(x, x+1, color='red', lw=2, linestyle='--')
axes.plot(x, x+2, color='orangered', lw=2, linestyle=':')
axes.plot(x, x+3, color='blue', lw=2, linestyle='-.')
axes.plot(x, x+4, color='blue', lw=2, marker='o', markerfacecolor='red')
#axes.plot(x, x**3, color='blue', lw=2, marker='x')
```

Plotting a contour

```python
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X,Y = np.meshgrid(x,y)
# Z1 = np.exp(-X**2 - Y**2)
# Z2 = np.exp(-(X-1)**2 - (y-1)**2)
# Z = (Z1 -Z2) * 2
Z = np.array([[x**-5]]]
print(X)
print(Y)
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = plt.figure(figsize=(14,6))
ax = fig.add_subplot(1,2,1, projection='3d')
ax.plot_surface(X,Y,Z, rstride=4,cstride=4, linewidth=0)
```
