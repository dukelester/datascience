Discretization and quantiling

Continuos values can be discretized using the cut() (bins based values) and qcut() (bins based on sample quantilies) functions::

```python
marks = np.random.randint(40, 100, 40)
marks
factor = pd.cut(marks, 10)
factor

arr = np.random.randn(30)
factor = pd.qcut(arr, [0,0.25,0.5,0.75,1])
factor
pd.value_counts(factor)

#passing infinite values to define the bins
arr = np.random.randn(20)
factor = pd.cut(arr, [-np.inf, 0, np.inf])
factor
pd.value_counts(factor)
```

Function Application



To apply a function to a pandas object , one should be aware o the three main methods::
These functions however depends on the main type of operation; this can be :
done on the dataframe, row-column or a series. These can be ::
The entire DataFrame, Series, row- or column-wise, or elementwise:

1. Tablewise Funtion Application: pipe()
2. Row or Column-wise Function Application: apply()
3. Aggregation API: agg() and transform()
4. Applying Elementwise Functions: applymap()





```python
#Tabular function application
'''
Dataframes and series can be passed into functions.If the function needs to be called in 
a chain then use the pipe() method.
'''

def extract_city_name(df):
    '''
    Chicago , IL ==> Chicago for city_name column
    '''
    df['city_name'] = df["city_and_code"].str.split(",").str.get(0)
    return df

def add_country_name(df, country_name =None):
    '''
    Chicago -> Chicago-US for city_name column
    '''
    col = "city_name"
    df["city_and_country"] = df[col] + country_name
    return df

df_p = pd.DataFrame({"city_and_code": ["Chicago, IL"]})
add_country_name(extract_city_name(df_p), country_name="US")
#Equivalent to

df_p.pipe(extract_city_name).pipe(add_country_name, country_name='US') #This is called method chaining using pipe()

```

The Aggregation API


This allows one to express more than one aggregation operations in a single consice way.
The entry point is the DataFrame.aggregate() or the alias DataFrame.agg()

```python
data

data.agg(np.sum)
data.aggregate(np.sum)
data.aggregate(np.cumsum)
data.agg('sum')
data.agg('prod')
data.agg('mean')
#one can pass multiple functions to the agg() as a list.
data.agg(['sum', 'mean', 'median', 'min', 'max'])
data['A'].agg(['sum', lambda x: x.mean()])


def mymean(x):
    return x.mean()
data['B'].agg(['sum', mymean])

#aggregating with a dict
data.agg({'A':'mean', 'B':'sum'})
data.dtypes
```

Transform API

The transform() method returns an object that is indexed the same (same axis) as the original.
Allows one to oprovide multiple operations at the same time rather than one-by-one.

```python
data
data.transform(np.abs)
data.transform([np.abs, lambda x : x + 1]) #using more than one function with transform ()
```
